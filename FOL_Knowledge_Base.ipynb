```python
import sys
sys.path.append("aima")
from aima.logic import *
```

```python
kb = FolKB()
```

```python
kb.tell(expr("LikesDessert(x, IceCream) ==> DreamsOf(x, Paris)"))
kb.tell(expr("Parent(x) & Parent(y) ==> SameDream(x, y)"))
kb.tell(expr("SameDream(x, y) & DreamsOf(x, d) ==> DreamsOf(y, d)"))
kb.tell(expr("CollectsCoins(x) ==> DreamsOf(x, CoinAlbum)"))
kb.tell(expr("PracticesBallet(x) ==> DreamsOf(x, SwanLake)"))
```

```python
kb.tell(expr("LikesDessert(Peter, IceCream)"))
kb.tell(expr("CollectsCoins(Peter)"))
kb.tell(expr("Parent(Mummy)"))
kb.tell(expr("Parent(Daddy)"))
kb.tell(expr("DreamsOf(Mummy, SeaTrip)"))
kb.tell(expr("PracticesBallet(Betty)"))
```

```python
print("Does Peter dream of Paris?")
print(kb.ask(expr("DreamsOf(Peter, Paris)")))
print("\nDoes Peter dream of Coin Album?")
print(kb.ask(expr("DreamsOf(Peter, CoinAlbum)")))
print("\nDo parents share the same dream?")
print(kb.ask(expr("SameDream(Daddy, Mummy)")))
print("\nDoes Betty dream of Swan Lake?")
print(kb.ask(expr("DreamsOf(Betty, SwanLake)")))
```
