```python

Person = {'Peter', 'Mummy', 'Daddy', 'Betty', 'AuntPolly'}
Dessert = {'Napoleon', 'Marmalade', 'Waffles', 'Marshmallows', 'IceCream'}
Dream = {'Paris', 'CoinAlbum', 'SwanLake', 'SeaTrip'}
Hobby = {'Yoga', 'Fishing', 'Sewing', 'Ballet', 'CoinCollecting'}
Day = {'Monday', 'Thursday'}
```

```python


kb = FolKB()

kb.tell(expr("Attends(Mummy, Yoga, Monday)"))
kb.tell(expr("Attends(Mummy, Yoga, Thursday)"))
kb.tell(expr("HasHobby(Mummy, Yoga)"))

kb.tell(expr("LikesDessert(x, IceCream) ==> DreamsOf(x, Paris)"))

kb.tell(expr("LikesDessert(Betty, Marmalade)"))

kb.tell(expr("LikesDessert(Mummy, Marshmallows)"))

kb.tell(expr("HasHobby(x, CoinCollecting) ==> DreamsOf(x, CoinAlbum)"))
kb.tell(expr("HasHobby(x, Ballet) ==> DreamsOf(x, SwanLake)"))
kb.tell(expr("HasHobby(AuntPolly, Sewing)"))


kb.tell(expr("HasHobby(Peter, Fishing)"))
kb.tell(expr("HasHobby(Daddy, Fishing)"))
kb.tell(expr("HasHobby(Peter, CoinCollecting)"))
kb.tell(expr("HasHobby(Betty, Ballet)"))


kb.tell(expr("LikesDessert(Peter, Marshmallows)"))
kb.tell(expr("LikesDessert(Peter, Marmalade)"))
kb.tell(expr("LikesDessert(Peter, Waffles)"))

kb.tell(expr("Parent(Mummy)"))
kb.tell(expr("Parent(Daddy)"))
kb.tell(expr("DreamsOf(Mummy, SeaTrip)"))
kb.tell(expr("Parent(Mummy) & Parent(Daddy) ==> SameDream(Mummy, Daddy)"))
kb.tell(expr("SameDream(x, y) & DreamsOf(x, d) ==> DreamsOf(y, d)"))

kb.tell(expr("LikesDessert(Daddy, NapoleonCake)"))
kb.tell(expr("LikesDessert(Betty, Marmalade)"))
kb.tell(expr("LikesDessert(Peter, Waffles)"))
kb.tell(expr("LikesDessert(AuntPolly, IceCream)"))


query1 = expr("LikesDessert(x, NapoleonCake)")
query2 = expr("DreamsOf(x, Paris)")

print("Query 1 Results:")
for r in fol_fc_ask(kb, query1):
    print(r)

print("\nQuery 2 Results:")
for r in fol_fc_ask(kb, query2):
    print(r)
```
